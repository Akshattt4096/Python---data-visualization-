## Question 3
Take a number from the user and check:

- Is it divisible by 5?
- Is it divisible by 3?

Print True or False.

```python
number = int(input("Enter a number: "))

if number % 5 == 0:
    print("Divisible by 5: True")
else:
    print("Divisible by 5: False")

if number % 3 == 0:
    print("Divisible by 3: True")
else:
    print("Divisible by 3: False")
```

## Question 4
Take a student's marks as input and print the grade:

Marks Range | Grade
------------|-------
90+ | A
75–89 | B
50–74 | C
Below 50 | Fail

```python
marks = int(input("Enter your marks: "))
if marks >= 90:
    print("Grade: A")
elif marks >= 75 and marks <= 89:
    print("Grade: B")
elif marks >= 50 and marks <= 74:
    print("Grade: C")
else:
    print("Grade: Fail")\
```

## Question 5
Take the age of a person and determine:

- Child (0–12)
- Teenager (13–19)
- Adult (20–59)
- Senior Citizen (60+)

```python
age = int(input("Enter your age: "))
if age >= 0 and age <= 12:
  print("Child")
elif age >= 13 and age <= 19:
  print("Teenager")
elif age >= 20 and age <= 59:
  print("Adult")
else:
  print("Senior Citizen")
```
